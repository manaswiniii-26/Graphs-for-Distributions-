```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=0
b=10

x=np.linspace(a-2,b+5,1000)
pdf=np.where((x>=a)&(x<=b),1/(b-a),0)

plt.figure(figsize=(8,5))
plt.plot(x,pdf,label='uniform distribution',color='orange')
plt.title('PDF of uniform distribution')
plt.xlabel('x')
plt.ylabel('pdf')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from scipy.stats import bernoulli

# Bernoulli distribution with p = 0.6
p = 0.6
x = [0, 1]
pdf = bernoulli.pmf(x, p)

plt.figure(figsize=(6,4))
plt.bar(x, pdf, width=0.2, color='blue', alpha=0.7)
plt.title('PDF of Bernoulli distribution (p=0.6)')
plt.xlabel('x')
plt.ylabel('Probability')
plt.grid(True, axis='y')
plt.show()
```
